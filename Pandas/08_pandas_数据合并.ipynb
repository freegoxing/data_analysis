```python
import numpy as np
import pandas as pd
```

```python
def make_df(index, columns):
    data = np.random.randint(0,10,size=(len(index),len(columns)))
    df = pd.DataFrame(data=data, index=index, columns=columns)
    return df
```

# 一，pd.concat()

```python
df1 = make_df([1,3,5],list('abcd'))
df2 = make_df([2,4,6],list('abed'))
display(df1,df2)
```

```python
#默认上下合并
pd.concat([df1,df2])
#左右合并
pd.concat([df1,df2], axis=1)
```

```python
#忽略行索引ignore_index
pd.concat([df1,df2],axis=1, ignore_index=True)
```

```python
#join参数
#外连接显示全部数据，会补nan-->outer默认
#内连接显示符合数据，不会补nan-->inner
```

```python
df3 = make_df([1,2,3],list('abc'))
df4 = make_df([2,3,4],list('bcd'))
display(df3,df4)
```

```python
pd.concat([df3,df4])
pd.concat([df3,df4],join='inner')
```

# pd.merge()

```python
#一对一
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df2 = pd.DataFrame({
    'id':[2,3,4],
    'sex':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df1,df2)
```

```python
#合并
pd.merge(df1,df2)
df1.merge(df2)
#根据相同合并
```

```python
#多对一
df3 = pd.DataFrame({
    'id':[1,2,2],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df4 = pd.DataFrame({
    'id':[2,3,4],
    'sex':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df3,df4)
```

```python
#df4中id为2的数据与df3中id为2 的数据合并
pd.merge(df3,df4)
```

```python
#多对多
df5 = pd.DataFrame({
    'id':[1,2,2],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df6 = pd.DataFrame({
    'id':[2,2,4],
    'sex':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df5,df6)
```

```python
#交叉合并
#类笛卡儿积
df5.merge(df6)
```

##     1,key规范化

        (1),on
            在多列名称相同时要完全相同才可
            那么要指定一个连接字段on

```python
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df2 = pd.DataFrame({
    'id':[2,3,4],
    'name':['bob','woman','man'],
    'job':['saler','writer','ceo']
})

display(df1,df2)
```

```python
df1.merge(df2,on='id')
#且前面的为x
df1.merge(df2, on='name')
```

        (2),left_on,right_on
            在左右都不相等时用left_on与right_on分别指定key
            df1.merge(df2)
            left_on是df1
            right_on是df2

```python
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df2 = pd.DataFrame({
    'id2':[2,3,4],
    'sex':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df1,df2)
```

```python
df1.merge(df2,left_on='id',right_on='id2')
```

        (2)left_index,right_index
            可以用行索引连接的字段

```python
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df2 = pd.DataFrame({
    'id2':[2,3,4],
    'sex':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df1,df2)
```

```python
df1.merge(df2,left_index=True,right_index=True)
df1.merge(df2,left_index=True,right_on='id2')
```

    2,how连接
      df1.merge(df2)
      内连接inner：显示符合数据，不会补nan（默认）
      外连接outer:显示全部数据，会补nan
      左连接left:显示df1全部数据与df2的公共数据
      右连接right:显示df2全部数据与df1的公共数据

```python
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df2 = pd.DataFrame({
    'id':[2,3,4],
    'sex':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df1,df2)
```

```python
df1.merge(df2)
df1.merge(df2,how='outer')
df1.merge(df2,how='left')
df1.merge(df2,how='right')
```

    3,suffixes列冲突
      如合并的列名相同默认变成'_x','_y'
      可以手动修改

```python
df1 = pd.DataFrame({
    'id':[1,2,3],
    'name':['div','lisa','bob'],
    'age':[23,34,54]
})

df2 = pd.DataFrame({
    'id':[2,3,4],
    'name':['man','woman','man'],
    'job':['saler','writer','ceo']
})

display(df1,df2)
```

```python
df1.merge(df2,on='id')
df1.merge(df2,on='id',suffixes=['_df1','_df2'])
```
