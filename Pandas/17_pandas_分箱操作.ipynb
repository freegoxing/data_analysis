```python
import numpy as np
import pandas as pd
import matplotlib as plot
```

分箱是连续数据的离散化    
有等距分箱与等频分箱

```python
columns = ['python','java','ruby','UI']
data = np.random.randint(60,100,size=(6,4))
df = pd.DataFrame(
    data=data,
    columns=columns
)
df
```

### 等距分箱

```python
#bins:分成几组
#     或添加断点
#right:默认左闭右开(True),False为左开右闭
s=pd.cut(df.iloc[:,0],bins=3)
s
```

```python
s.value_counts().plot.bar()
```

```python
#bins:分成几组
#     或添加断点
#right:默认左闭右开(True),False为左开右闭
#labels:分类标签
s2 = pd.cut(
    df.iloc[:,0],
    bins=[60,90,120,150],
    labels=['C','B','A']
)
s2.value_counts().plot.bar()
```

### 等频分箱

```python
#q:几等分
s2 = pd.cut(
    df.iloc[:,0],
    bins=3,
    labels=['C','B','A']
)
s2.value_counts().plot.bar()
```
