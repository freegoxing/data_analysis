```python
import numpy as np
import pandas as pd
import openpyxl                        #excel数据需要
from sqlalchemy import create_engine   #数据库需要
import pymysql
```

# 一,CSV数据

```python
data = np.random.randint(60,100,size=(10,5))
columns = ['python','MySQL','JS','java','numpy']
df = pd.DataFrame(
    data=data,
    columns=columns
)
df
```

### df.to_csv()保存成csv数据

```python
#path_or_buf:路径
#sep:分隔符(一般为‘，’)
#header:是否保存列索引
#index:是否保存行索引
df.to_csv('test.csv',sep=',',header=True,index=True)
```

### pd.read_csv()加载csv数据

```python
#sep:分隔符(一般为‘，’)
#header=None:不获取列的名字
#header=[n1，n2...]:表示从0开始n1,n2..作为列索引
pd.read_csv('test.csv',sep=',',index_col=0)
```

### pd.read_table()加载csv数据

```python
#与read_csv类似
#sep默认为\t
pd.read_table('test.csv',sep=',',index_col=0)
```

# 二,EXCEL数据

```python
data = np.random.randint(60,100,size=(10,5))
columns = ['python','MySQL','JS','java','numpy']
df = pd.DataFrame(
    data=data,
    columns=columns
)
df
```

### df.to_excel()保存成excel数据

```python
#sheet_nam:工作表名称
#header:是否保存列索引
#index:是否保存行索引
df.to_excel('test.xlsx',sheet_name='sheet1',header=True,index=False)
```

### pd.read_excel()加载excel数据

```python
#header=[n1，n2...]:表示从0开始n1,n2..作为列索引
#sheet_name=0:表示从0开始读取第1个
#names:替换原本的columns
pd.read_excel('test.xlsx',sheet_name='sheet1',names=list('asdfg'))
```

# 三,MySQL数据

```python
data = np.random.randint(0,150,size=(150,3))
columns = ['python','pandas','numpy']
df = pd.DataFrame(
    data=data,
    columns=columns
)
df
```

先连接MySQl

```python
#mysql+pymysql:数据库类型+驱动
#root:qwer1234QWER:用户名+密码
#R@localhost:3306:数据库地址+端口
#pandas数据库名
conn = create_engine('mysql+pymysql://root:qwer1234QWER@localhost:3306/pandas')
```

### df.to_sql()保存成db数据

```python
df.to_sql(
    name='score',        #数据库中表的名字
    con=conn,            #数据库连接对象
    index=False,         #是否保留行索引
    if_exists='append'   #如果表存在，则追加数据
)
```

### df.read_sql()读取db数据

```python
#Mysql语法,要实现的功能
sql = 'select * from score'
```

```python
pd.read_sql(
    sql=sql,
    con=conn
    # index_col='python'  指定行索引的类名
)
```
