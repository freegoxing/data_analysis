```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

kind : str           
    The kind of plot to produce:          
           
折线图    - 'line' : line plot (default)       
柱状图    - 'bar' : vertical bar plot         
条形图    - 'barh' : horizontal bar plot         
直方图    - 'hist' : histogram         
箱型图    - 'box' : boxplot            
核密度估计    - 'kde' : Kernel Density Estimation plot          
频率    - 'density' : same as 'kde'          
面积图    - 'area' : area plot          
饼图    - 'pie' : pie plot              
散点图    - 'scatter' : scatter plot (DataFrame only)         
    - 'hexbin' : hexbin plot (DataFrame only)        
ax : matplotlib axes object, default None   

## 一，折线图

#### series图表

```python
s = pd.Series([12,54,22,6,87])
s.plot()
```

#### 正弦曲线

```python
#点之间的间隔要小
x = np.arange(0,2*np.pi,0.01)
y = np.sin(x)
s = pd.Series(data=y, index=x)
s.plot()
```

#### dataframe图表

```python
data = np.random.randint(60,100,size=(6,4))
index = ['1st','2nd','3rd','4th','5th','6th']
columns = ['python','numpy','pandas','java']
df = pd.DataFrame(data=data,index=index,columns=columns)
df
```

```python
#默认每一列一条线
df.plot()
#利用.T变成每一行一条线
df.T.plot()
```

## 二,条形图与柱形图

#### dataframe

```python
df = pd.DataFrame(np.random.randint(34,76,size=(5,8)))
df
```

```python
#stacked=True同一位置的数据堆叠
#barh图表旋转了90°
df.plot(kind='bar')
# df.plot.bar()
df.plot.bar(stacked=True)
df.plot.barh()
```

#### series

```python
s = pd.Series(data=[123,234,567,123])
s.index = ['python','numpy','java','pandas']
s.plot.bar()
```

## 三,直方图

```python
#hist直方图
#bins:组数
#density:是否把频数转变为频率(默认False)
s = pd.Series([1,22,33,22,22,44,55,33])
s.plot(kind='hist',bins=5,density=True)
```

```python
#kde图:核密度估计
s.plot(kind='hist',bins=5,density=True)
s.plot(kind='kde')
```

## 四,饼图

```python
df = pd.DataFrame(
    data = np.random.randint(60,100,size=(3,4)),
    index = list('ABC'),
    columns = ['python','java','UI','pandas']
)
df
```

```python
#subplots显示子图
#figsize大小
#autopct='%.1f%%':.1f%表示一位小数
df.plot(kind='pie',subplots=True,figsize=(16,16),autopct='%.1f%%')
```

## 五,散点图

```python
data = np.random.normal(size=(1000,2))
df = pd.DataFrame(data=data,columns=list('AB'))
df
```

```python
#x='A',y='B'使用A为x轴B为y轴
df.plot(kind='scatter',x='A',y='B')
```

## 六,面积图

```python
df = pd.DataFrame(data=np.random.rand(10,4),columns=list('ABCD'))
df
```

```python
#stacked=True同一位置的数据堆叠
df.plot(kind='area',stacked=False)
```

## 七,箱型图

```python
df = pd.DataFrame(data=np.random.rand(10,4),columns=list('ABCD'))
df
df.plot(kind='box')
```
