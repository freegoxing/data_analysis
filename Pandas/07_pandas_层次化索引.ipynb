```python
import numpy as np
import pandas as pd
```

一，创建多层次化索引

    1隐式构造

```python
#dataframe索引传递多个数组
data = np.random.randint(60,101,size=(6,6))

index = [
    ['一班','一班','一班','二班','二班','二班'],
    ['小明','小红','李华','张三','李四','王五']
]

columns = [
    ['期中','期中','期中','期末','期末','期末'],
    ['语文','数学','英语','语文','数学','英语']
]

df = pd.DataFrame(data=data, index=index, columns=columns)
df
```

```python
#sereies创建
data = np.random.randint(60,101,size=(6,))

index = [
    ['一班','一班','一班','二班','二班','二班'],
    ['小明','小红','李华','张三','李四','王五']
]

s = pd.Series(data=data,index=index)
s
```

    2,显示构造pd.Multiindex

```python
#使用数组
data = np.random.randint(60,101,size=(6,6))

index = pd.MultiIndex.from_arrays([
    ['一班','一班','一班','二班','二班','二班'],
    ['小明','小红','李华','张三','李四','王五']
])

columns = [
    ['期中','期中','期中','期末','期末','期末'],
    ['语文','数学','英语','语文','数学','英语']
]

df = pd.DataFrame(data=data, index=index, columns=columns)
df
```

```python
#使用tuple(元组)
data = np.random.randint(60,101,size=(6,6))

index = pd.MultiIndex.from_tuples(
    (
        ('一班','小明'),('一班','小红'),('一班','李华'),
        ('二班','张三'),('二班','李四'),('二班','王五')
    )
)

columns = [
    ['期中','期中','期中','期末','期末','期末'],
    ['语文','数学','英语','语文','数学','英语']
]

df = pd.DataFrame(data=data, index=index, columns=columns)
df
```

```python
#笛卡尔积
#A:(a,b),B:(c,d)
#A*B=(a,c),(a,d),(b,c),(b,d)
data = np.random.randint(60,101,size=(6,6))

index = pd.MultiIndex.from_product([
    ['一班','二班'],
    ['小明','小红','李华']
])

columns = [
    ['期中','期中','期中','期末','期末','期末'],
    ['语文','数学','英语','语文','数学','英语']
]

df = pd.DataFrame(data=data, index=index, columns=columns)
df
```

二，索引与切片

    1,series多层结构

```python
s
```

```python
#显式索引
#第一层
s['一班']
s.loc['一班']
s.loc[['一班']]#多层结构保留了一班
s.loc[['一班','二班']]
# #第二层要通过第一层
s['一班']['小红']
s.loc['一班','小红']

#隐式索引
#直接对values无层次
s[0]
s.iloc[0]
```

```python
#显式切片左闭右闭
s['一班':'二班']
s.loc['一班':'二班']

#隐式索引左闭右开
#只看数字下标
s[0:4]
s.iloc[0:4]
```

    2,dataframe多层结构

```python
df
```

```python
#获取元素
df['期中']['数学'][0]

df.iloc[0,1]
df.loc[('一班','小红'),('期中','英语')]
```

```python
#列索引
df['期中']['语文']
df['期中','语文']

df.iloc[:,2]
df.iloc[:,[0,2,4]]
df.loc[:,('期中','数学')]


#行索引
df.loc['一班']
df.loc['一班'].loc['李华']
df.loc['一班','李华']
df.loc[('一班', '李华')]
```

```python
#行切片
df.iloc[1:5]
df.loc[('一班','小红'):('二班','王五')]
df.loc['一班':'二班']

#列切片
df.iloc[:,1:5]
df.loc[:,'期中':'期末']

#切片建议用iloc隐式索引
```

三，索引的堆叠

```python
df
```

```python
#stack将行与列交换
df.stack()#默认是最底层的
#利用参数level改变
df.stack(level=0)
#多层索引由外到内标签由0增加或反方向由-1减少
```

```python
#unstack将行与列交换
df.unstack(level=0)
#多层索引由外到内标签由0增加或反方向由-1减少
df.unstack(level=0, fill_value=0)
```
