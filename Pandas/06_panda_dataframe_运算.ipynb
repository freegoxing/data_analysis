```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(
    data=np.random.randint(60,101,size=(3,6)),
    index=['小明','小刘','李华'],
    columns=['语文','数学','英语','物理','化学','生物']
)

df2 = pd.DataFrame(
    data=np.random.randint(60,101,size=(3,6)),
    index=['小明','小刘','李华'],
    columns=['语文','数学','英语','物理','化学','生物']
)

display(df1,df2)
```

在values为数值时与numpy一致

```python
s1 = df1 + 10 #加+
s2 = df1 - 10 #减-
s3 = df1 * 10 #乘*
s4 = df1 / 10 #除/
s5 = df1 // 2 #整除//(求商)
s6 = df1 ** 2 #次方**
s7 = df1 % 3 #求模%
display(s1,s2,s3,s4,s5,s6,s7)
```

dataframe间的运算

```python
#对应元素
m1 = df1+ df2
m2 = df1 - df2
m3 = df1*df2
display(m1,m2,m3)
```

```python
df3 = pd.DataFrame(
    data=np.random.randint(60,101,size=(4,5)),
    index=['小明','小刘','李华','小红'],
    columns=['语文','数学','英语','物理','化学']
)
display(df1,df3)
```

```python
#无广播机制
df1+df3
```

```python
#利用add填充
df3.add(df1,fill_value=0)
```

series与dataframe间的运算

```python
s = pd.Series(np.random.randint(61,100,size=(6,)),index=df1.columns)
display(df1,s)
```

```python
#对应标签军加上series的values
df1+s

df1.add(s)
#利用add参数axis决定相加的维度
#0(或index)表示列
#1(或columns)表示行
```

```python
s1 = pd.Series([13,43,65],index=df1.index)
s1
```

```python
df1.add(s1,axis=0)
```
