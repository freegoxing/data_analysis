```python
import numpy as np
import pandas as pd
```

```python
index = ['bob','lisa','jen','lihua']
columns = ['python','java','ruby','UI']
data = np.random.randint(60,100,size=(4,4))
df = pd.DataFrame(
    data=data,
    index=index,
    columns=columns
)
df
```

## 一,无放回抽样

```python
#按行取其数字下标
df.take([0,3,1])
df.take([1,0],axis=1)
```

```python
#随机排序
#把给定的值随机排序
np.random.permutation([0,1,2,3])

df.take(np.random.permutation([0,2,3]))
```

## 二,有放回抽样

```python
#可能有重复值
np.random.randint(0,3,size=(10,))

df.take(np.random.randint(0,3,size=(10,)))
```
