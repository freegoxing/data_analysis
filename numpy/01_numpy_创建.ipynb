    1,创建一所有元素为1（0）的多维数组

    3,创建一个二维的单位矩阵
      np.eye(N, M, k=0, dtype=float)
      N:行
      M:列
      k:向右偏移的位置(负数向左)

```python
n1 = np.eye(6, 4)
n2 = np.eye(6, 4, k=2)
n3 = np.eye(6, 4, k=-2)
display(n1, n2, n3)
```

    4,创建一个等差数列

      start:开始值

      num:数列个数
      endpoint:是否包含结束值
      retstrp:是否返回步长

```python
n= np.linspace(0,99, num=100)
n
```

    5,创建一个数值范围的数组

      start:开始值（可选，不选为0）
      stop:结束值（不包含）
      step:步长（可选，不选为1）

```python
n = np.arange(1,89,7)
n
```

    6,创建随机数（random模块）
        (1),创建随机多维数组
            np.random.randint(low, high=None, size=None, dtype=None)
            low:最小值
            high:最大值
                *high=None在【0, low)区间
                *high!=None在【low， high）区间
            size:数组形状(默认为一个值)

```python
n1 = np.random.randint(3, 7)
n2 = np.random.randint(3, 7, size=(3,4))
display(n1, n2)
```

```python
n = np.random.randint(256, size=(100, 200, 3))
plt.imshow(n)
```
