```python
#一维与列表完全一致
l = [23,354,7,76,8.23,25,4,8678,97]
l1 = l[2:6]
l2 = l[::-1]
n = np.array(l)
n1 = n[2:6]
n2 = n[::-1]
display(l1,l2,n1,n2)
```

```python
#取一列
n1 = n[:,0]#取所有行的第一个元素
#取连续的列
n2 = n[:,2:5]#取所有行的第3到5个元素
#取不连续的多行(多个中括号)
n3 = n[:,[2,6]]
display(n1,n2,n3)
```

三，函数

    1，变形reshape()

```python
n = np.arange(1,25)
display(n, n.shape)
```

```python
#变成二维
n2 = np.reshape(n, (4,6))

#变成一维
n3 = n2.reshape(24)
display(n2, n2.shape, n3)
```

```python
#利用-1表示剩余长度
n3.reshape((-1,4))
```

    2,级联concatenate()

```python
n1 = np.random.randint(0,100,size=(3,5))
n2 = np.random.randint(0,100,size=(3,5))
display(n1,n2)
```

```python
#默认上下合并(列数相同)
#利用axis参数 0表示第一个维度行
#            1表示第一个维度列       
n3 = np.concatenate((n1,n2))
n4 = np.concatenate((n1,n2), axis=1)
display(n3,n4)
```

    3,水平（垂直）合并np.hstack(np.vstack)

```python
n5 = np.hstack((n1,n2))
n6 = np.vstack((n1,n2))
display(n5,n6)
```

    4,拆分split,hsplit,vsplit

```python
n = np.random.randint(0,100,size=(6,4))
n
```

```python
#垂直拆分
#均分
n1 = np.vsplit(n,3)
#不均分
#括号中为对应垂直索引前分割
n2 = np.vsplit(n,(1,2,4))
display(n1,n2)

#hsplit同理
```

```python
#split均可（默认按行）
#利用axis参数
n1 = np.split(n, 2, axis=0)
n2 = np.split(n, 2, axis=1)
display(n1, n2)
```
