```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.random.randint(0,11,size=60)
s = pd.Series(x).value_counts()
s
```

```python
#表示每一个数出现的概数
#bins:组数
#     或人为添加断点
#facecolor:填充的颜色
#edgecolor:边框的颜色
#density概率分布
# plt.hist(x,bins=5)
plt.hist(
    x,
    bins=[0,2,5,8,10],
    facecolor='m',
    edgecolor='k',
    density=True
)

plt.xticks(range(10))
plt.show()
```
