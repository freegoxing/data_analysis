```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']='SimHei'
```

## 饼图

```python
x = [10,20,48,38]
labels = ['python','java','pycharm','vscode']

plt.pie(
    x,
    autopct='%.0f%%',     #显示百分比,（几位小数）
    pctdistance=0.8,      #百分比的位置
    labels=labels,        #标签
    labeldistance=1.4,    #标签的位置
    textprops={           #文字样式
        'fontsize':12,
        'color':'m'
    },
    explode=[0,0.5,0,0] #分裂效果,值越大越突出
)
plt.show()
```

## 甜甜圈图

```python
x = [10,20,48,38]
labels = ['python','java','pycharm','vscode']

plt.pie(
    x,
    autopct='%.0f%%',     
    pctdistance=0.8,      
    labels=labels,        
    labeldistance=1.4,    
    textprops={           
        'fontsize':12,
        'color':'m'
    },
    wedgeprops={           #甜甜圈图
        'width':0.6,       #空心圆的剩余半径
        'edgecolor':'k'    #边框的颜色
    }
)
plt.show()
```
