```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 读取图片

```python
img = plt.imread('url.jpg')
img
```

## 显示图片

```python
plt.imshow(img)
```

#### 垂直翻转

```python
# plt.imshow(img,origin='lower')
plt.imshow(img[::-1])
```

#### 水平翻转

```python
plt.imshow(img[:,::-1])
```

#### 截取图片

```python
plt.imshow(img[75:150,50:150])
```

## 保存图片

```python
img2 = img[75:150,50:150]
img2.shape
```

```python
plt.imsave('cat_face.jpg',img2)
```
