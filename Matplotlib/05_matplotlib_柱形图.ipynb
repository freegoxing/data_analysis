```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_excel('test.xlsx')
x = np.arange(0,10)
y1, y2, y3, y4, y5 = df.iloc[:,0], df.iloc[:,1], df.iloc[:,2], df.iloc[:,3], df.iloc[:,4]
df
```

```python
#width柱子的宽度
plt.bar(x,y1,width=0.5)
```

## 多个图

#### 簇状柱形图

```python
w=0.2
plt.bar(x-w,y1,width=w,label='python')
plt.bar(x,y2,width=w,label='MySQL')
plt.bar(x+w,y2,width=w,label='JS')

plt.legend()
```

## 堆叠柱状图

```python
#bottom起始的位置
plt.bar(x,y1,label='python')
plt.bar(x,y2,label='MySQL',bottom=y1)
plt.bar(x,y2,label='JS',bottom=y2+y1)

plt.legend()
```

## 条形图

```python
plt.barh(x,y1)
```
