```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
```

## 折线图

```python
plt.figure(figsize=(6,4))

x = np.array(['python','java','UI','scrapy','urllib'])
y = np.random.randint(60.100,size=(5,))

plt.plot(x,y)
```

#### 多条线

```python
plt.figure(figsize=(6,5))

x = np.random.randint(0,10,size=15)

plt.plot(x)
plt.plot(x.cumsum())
```

#### 读取excel数据

```python
df = pd.read_excel('test.xlsx')
df
```

```python
x = np.arange(0,10)
y1, y2, y3, y4, y5 = df.iloc[:,0], df.iloc[:,1], df.iloc[:,2], df.iloc[:,3], df.iloc[:,4]
```

```python
plt.figure(figsize=(6,5))

plt.plot(x,y1,label='python')
plt.plot(x,y2,label='MySQL')
plt.plot(x,y3,label='JS')
plt.plot(x,y4,label='java')
plt.plot(x,y5,label='numpy')

plt.legend()
```

```python
#width柱子的宽度
plt.bar(x,y1,width=0.5)
```

#### 多个图

###### 簇状柱形图

```python
w=0.2
plt.bar(x-w,y1,width=w)
plt.bar(x,y2,width=w)
plt.bar(x+w,y2,width=w)
```
