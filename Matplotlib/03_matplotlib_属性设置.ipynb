```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
```

## 绘图基本属性

#### 图例

###### legend 

```python
plt.figure(figsize=(6,4))

x = np.linspace(0,2*np.pi,1000)
plt.plot(x,np.sin(x),label='sin')
plt.plot(x,np.cos(x),label='cos')

#图例
plt.legend()
```

```python
plt.figure(figsize=(6,4))

x = np.linspace(0,2*np.pi,1000)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))

#默认按图表创建顺序显示图例
#fontsize:图例大小
#loc:图例位置(默认为'best')
#ncol:显示几列(默认为1)
plt.legend(
    ['Sin','Cos'],
    fontsize=10,
    loc='center right',
    ncol=2
)
```

#### 线条

###### c(color):颜色(01有)
###### ls(linestyle):样式(01有)
###### lw(linewidth):线宽
###### label:标签
###### apha:透明度(0~1间)
###### marker:标记
###### mfc(marker of color):标记的背景颜色
###### ms(markersize):标记的大小
###### markeredgecolor:标记边缘的颜色
###### markeredgewidth:标记边缘的宽度

```python
# **Markers**

# =============   ===============================
# character       description
# =============   ===============================
# ``'.'``         point marker
# ``','``         pixel marker
# ``'o'``         circle marker
# ``'v'``         triangle_down marker
# ``'^'``         triangle_up marker
# ``'<'``         triangle_left marker
# ``'>'``         triangle_right marker
# ``'1'``         tri_down marker
# ``'2'``         tri_up marker
# ``'3'``         tri_left marker
# ``'4'``         tri_right marker
# ``'8'``         octagon marker
# ``'s'``         square marker
# ``'p'``         pentagon marker
# ``'P'``         plus (filled) marker
# ``'*'``         star marker
# ``'h'``         hexagon1 marker
# ``'H'``         hexagon2 marker
# ``'+'``         plus marker
# ``'x'``         x marker
# ``'X'``         x (filled) marker
# ``'D'``         diamond marker
# ``'d'``         thin_diamond marker
# ``'|'``         vline marker
# ``'_'``         hline marker
```

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,10)
plt.plot(
    x,np.sin(x),
    marker='*',
    c='m',ls='-.',
    label='sin',
    mfc='c',
    lw=3,
    markersize=15,
    markeredgecolor='y',
    markeredgewidth=1.5,
    alpha=0.7
)


plt.legend()
```

#### 坐标轴

###### xticks:x坐标轴刻度
###### yticks:y坐标轴刻度

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

#设置坐标值
#fontsize:坐标轴刻度的大小
#color:坐标轴刻度的颜色
#label:坐标轴刻度的标签
#ha:坐标轴刻度的对齐方式(默认右对齐)
plt.xticks(
    ticks=np.arange(0,2*np.pi,0.25*np.pi),
    fontsize=12.5,
    color='m'
)

plt.yticks(
    ticks=[-1,0,1],
    labels=['min',0,'max'],
    ha='left'
)

plt.show
```

###### xlim:x坐标轴范围
###### ylim:y坐标轴范围

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

plt.xlim(-2,8)
plt.ylim(-2,2)
```

###### axis()

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

#设置坐标轴范围([xmin,xmax,ymin,ymax])
plt.axis([-2,8,-2,2])

#option
#on:显示范围->以元组的方式
#off:不显示坐标轴
#equal:x与y单位距离相等
#scaled:自动缩放坐标轴与图片匹配
#tight:自动紧凑
#square:让画布呈现为正方形

plt.axis('equal')
```

#### 标题

###### title()与suptitle()

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

#标题
plt.title(
    label='sin曲线',
    fontsize=20,
    loc='right'
)

#f父标题
plt.suptitle(
    '三角曲线',
    fontsize=40,
    y=1
)
```

#### 网格

###### grid()

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

#axis:显示那个轴
plt.grid(
    ls='--',
    color='g',
    lw=1,
    axis='y'
)
```

#### 标签

###### xlabel
###### ylabel

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

#rotatio旋转角度(y轴默认旋转90°)
plt.xlabel(
    'y=sinx',
    fontsize=20,
    rotation=45
)

#horizontalalignment:水平对齐方式
plt.ylabel(
    'y=sinx',
    fontsize=20,
    rotation=0,
    horizontalalignment='right'
)
```

#### 文本

###### text

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,11,10)
y = np.array([32,43,45,56,1,35,56,98,34,25])
plt.plot(x,y)

#画文字
#位置为坐标轴对应的刻度
#ha:水平轴向对齐方式
#va:竖平轴向对齐方式
for a,b in zip(x,y):
    plt.text(
        x=a,
        y=b,
        s=b,
        fontsize=20,
        ha='center',
        va='bottom'
    )
```

#### 注释

###### annotate

```python
plt.figure(figsize=(6,5))

x = np.linspace(0,9,10)
y = np.array([32,43,45,56,1,35,56,98,34,25])
plt.plot(x,y)

#标注
plt.annotate(
    text='max',          #xy:arrow指向的位置
    xy=(7,98),           #xytext:文本的位置
    xytext=(4,88),
    #箭头
    arrowprops={
        'width':2,       #width:arrow线宽的大小
        'headwidth':16 , #headwidth:arrow头部的大小
        'facecolor':'m'
    }
)
```

#### 保存图片

###### savefig()

```python
fig=plt.figure(figsize=(6,5))

x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))

# plt.savefig('sin.png')

fig.savefig(
    fname='sin2.png',
    dpi=125,
    facecolor='m'
)
```
