```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

## 箱型图

```python
x = [-10,1,2,3,5,7,9,0,34]

plt.boxplot(x)
plt.show()
```

## 多个箱型图

```python
x1 = np.random.randint(0,100,size=10)
x2 = np.random.randint(0,100,size=10)
x3 = np.random.randint(0,100,size=10)

plt.boxplot([x1,x2,x3])
plt.show()
```

```python
data = np.random.normal(size=(500,4))
labels=['A','B','C','D']

plt.boxplot(
    data,
    labels=labels,          #标签
    notch=True,             #箱型图样式
    sym='m*'                #异常值的颜色与标记
)

plt.show()
```
