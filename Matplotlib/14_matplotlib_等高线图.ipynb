## 等高线图

```python
a = np.array([2,3,6])
b = np.array([7,3,6,3])

A,B = np.meshgrid(a,b)
display(A,B)
```

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

#把xy变成网格
X,Y = np.meshgrid(x,y)

#Z
Z = np.sin(X*Y)

#画等高线
cp = plt.contourf(X,Y,Z)

#颜色条
plt.colorbar()
```
