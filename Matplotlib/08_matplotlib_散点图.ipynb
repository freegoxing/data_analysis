```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

## 散点图

```python
x = range(1,10,1)
y = range(40,130,10)

plt.scatter(x,y)
```

## 气泡图

```python
data = np.random.randn(100,2)
s = np.random.randint(50,200,size=100)
c = np.random.randn(100)

plt.scatter(
    data[:,0],
    data[:,1],
    s=s,
    c=c
)
```

## 六边形散点图

```python
data = np.random.randn(100,2)

plt.hexbin(
    data[:,0],
    data[:,1],
    gridsize=20,       #六边形点的大小
    cmap='rainbow'     #色彩
)
```
