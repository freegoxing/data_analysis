## 基本绘图

###### plt.plot()

```python
#抛物线
x = np.linspace(-50, 50,101)
y = x**2

#画图:折线图
#默认第一个位置为x轴,第二个位置为y轴
plt.plot(y,x)
```

#### 样式与颜色

```python
x = np.linspace(-50, 50,101)
y = x**2
#c=颜色(或color)
#b(blue),g(green),r(red),c(cyan青),m(magenta品红),y(yellow),k(black),w(white)

#ls=line style(线的样式)
#-(solid line style),--(dashed line style),-.(dash-dot line style),:(dotted line style)

# plt.plot(y,x,c='m',ls='-.')
plt.plot(y,x,'m-.')
```

## 画布配置

###### plt.figure()

```python
#figsize画布的大小,宽高
#dpi分辨率
#facecolor:背景颜色
plt.figure(figsize=(2*np.pi,2),dpi=100,facecolor='r')

#画正弦
x = np.linspace(0,2*np.pi)
y = np.sin(x)
plt.plot(x,y,'c')

#设置网格
plt.grid()
```

#### 在一个画布上画多个图

```python
plt.figure(figsize=(10,10))
x = np.linspace(0,2*np.pi)

plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.plot(x,np.arcsin(x))
```

```python
#show
#分隔
plt.figure(figsize=(10,10))
x = np.linspace(0,2*np.pi)

plt.plot(x,np.sin(x))
plt.show()

plt.plot(x,np.cos(x))
plt.show()
plt.plot(x,np.arcsin(x))
```
