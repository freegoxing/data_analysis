```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
```

## 均匀分布

#### subplot()

```python
#2行2列
fig = plt.figure(figsize=(5,3))

x = np.linspace(-2*np.pi,2*np.pi,30)

#子图1
ax1 = plt.subplot(2,2,1)#2行2列中第一个
ax1.plot(x,np.sin(x))
ax1.set_title('子图1')

#子图2
ax2 = plt.subplot(2,2,2)#2行2列中第二个
ax2.plot(x,np.cos(x))
ax2.set_title('子图2')

#子图3
ax3 = plt.subplot(2,2,3)#2行2列中第三个
ax3.plot(x,np.sin(x*2))
ax3.set_title('子图3')

#子图4
ax4 = plt.subplot(2,2,4)#2行2列中第四个
ax4.plot(x,np.cos(x*2))
ax4.set_title('子图4')

#自动调整布局
fig.tight_layout()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,30)

#子图1
ax1 = plt.subplot(2,2,1)#2行2列中第1个
ax1.plot(x,np.sin(x))
ax1.set_title('子图1')

#子图2
ax2 = plt.subplot(2,2,2)#2行2列中第2个
ax2.plot(x,np.cos(x))
ax2.set_title('子图2')

#子图3
ax3 = plt.subplot(2,1,2)#2行一列中第2个
ax3.plot(x,np.sin(x*2))
ax3.set_title('子图3')
```

#### subplots()

```python
x = np.linspace(-2*np.pi,2*np.pi,30)

#3行3列
fig,ax = plt.subplots(3,3)
ax1,ax2,ax3 = ax
ax11,ax12,ax13 = ax1
ax21,ax22,ax23 = ax2
ax31,ax32,ax33 = ax3

#设置画布大小
fig.set_figwidth(8)
fig.set_figheight(5)

#第1行
ax11.plot(x,np.sin(x))
ax12.plot(x,np.cos(x))
ax13.plot(x,np.sin(x*2))

#第2行
ax21.plot(x,np.exp(x))
ax22.plot(x,np.log(x))
ax23.plot(x,x*2)

#第1行
ax31.plot(x,np.cos(np.sin(x)))
ax32.plot(x,np.sin(np.cos(x)))
ax33.plot(x,np.sin(np.cos(x*2)))

plt.tight_layout()
```

## 图形嵌套

#### add_subplot()

```python
fig = plt.figure(figsize=(5,3))

x = np.linspace(-5,5,100)

#子图1
axes1 = fig.add_subplot(1,1,1)
axes1.plot(x, np.tan(x))

#嵌套一个子图2
axes2 = fig.add_subplot(3,3,9,facecolor='pink')#把子图1分成3*3的小区域子图2在第一个区域
axes2.plot(x, np.sin(x))
```

#### axes()与add_axes()

```python
fig = plt.figure(figsize=(8,5))

x = np.linspace(-5,5,100)

#图一
plt.plot(x,np.sin(x))

#嵌套图1
#[left,bottom,width,height]
#原点的x位置,原点y位置,图宽，图高
axes1 = plt.axes([0.3,0.65,0.2,0.2])
axes1.plot(x,x*2,'g:')

#嵌套图2
axes2 = fig.add_axes([0.50,0.15,0.2,0.2])
axes2.plot(x,x**2,'m-.')
```

## 双轴显示

```python
fig = plt.figure(figsize=(8,5))

x = np.linspace(0,10,1000)

#图1
#gca()得到当前的轴
axes1 = plt.gca()
axes1.plot(x,np.exp(x))
axes1.set_xlabel('time')
axes1.set_ylabel('exp',c='b')
axes1.tick_params(axis='y',labelcolor='blue')#刻度的设置

#图2
#twinx()共享x轴
axes2 = axes1.twinx()
axes2.plot(x,np.sin(x),'r')
axes2.set_ylabel('sin',c='r')
axes2.tick_params(axis='y',labelcolor='red')
```
