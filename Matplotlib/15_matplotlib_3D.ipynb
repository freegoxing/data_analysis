```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']='SimHei'
```

```python
#3D引擎
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

## 3D图

#### 3D折线图

```python
fig = plt.figure(figsize=(5,6))

x = np.linspace(0,100,1010)
y = np.sin(x)
z = np.cos(x) 

#3D折线图
axes = Axes3D(fig,auto_add_to_figure=False)
fig.add_axes(axes)
axes.plot(x,y,z)

axes.set_xlabel('X',fontsize=15)
axes.set_ylabel('sin(X)',fontsize=15)
axes.set_zlabel('cos(X)',fontsize=15)

plt.show()
```

#### 3D散点图

```python
fig = plt.figure(figsize=(5,3))
axes = Axes3D(fig,auto_add_to_figure=False)
fig.add_axes(axes)

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

axes.scatter(x,y,z)
```

#### 3D柱形图

```python
fig = plt.figure(figsize=(8,10))
axes = Axes3D(fig,auto_add_to_figure=False)
fig.add_axes(axes)

x = np.arange(1,5)

for i in x:
    axes.bar(
        np.arange(4),                       #y轴
        np.random.randint(10,100,size=(4)), #z轴
        zs=i,                               #在x轴上的第几个
        zdir='x',                           #在哪个轴上排列
        width=0.5 ,                         #柱子的宽度
        alpha=0.5
    )

axes.set_xlabel('x',fontsize=15)
axes.set_ylabel('y',fontsize=15)
axes.set_zlabel('z',fontsize=15)
```
